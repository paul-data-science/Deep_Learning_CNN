```python
import numpy as np
from PIL import Image
import os
import cv2
```

```python
data = []
labels = []

cats=os.listdir("C:\\Users\\win_10\\Desktop\\Final Project\\animals\\cats")
for cat in cats:
    imag=cv2.imread("C:\\Users\\win_10\\Desktop\\Final Project\\animals\\cats\\"+cat)
    cat_array = np.asarray(imag)
    img_from_ar = Image.fromarray(cat_array, 'RGB')
    resized_image = img_from_ar.resize((50, 50))
    data.append(np.array(resized_image))
    labels.append(0)
    
```

```python
bats=os.listdir("C:\\Users\\win_10\\Desktop\\Final Project\\animals\\bats")
for bat in bats:
    imag=cv2.imread("C:\\Users\\win_10\\Desktop\\Final Project\\animals\\bats\\"+bat)
    bat_array = np.asarray(imag)
    img_from_ar = Image.fromarray(bat_array, 'RGB')
    resized_image = img_from_ar.resize((50, 50))
    data.append(np.array(resized_image))
    labels.append(1)

spiders=os.listdir("C:\\Users\\win_10\\Desktop\\Final Project\\animals\\spiders")
for spider in spiders:
    imag=cv2.imread("C:\\Users\\win_10\\Desktop\\Final Project\\animals\\spiders\\"+spider)
    sp_array = np.asarray(imag)
    img_from_ar = Image.fromarray(sp_array, 'RGB')
    resized_image = img_from_ar.resize((50, 50))
    data.append(np.array(resized_image))
    labels.append(2)
    
wolverines=os.listdir("C:\\Users\\win_10\\Desktop\\Final Project\\animals\\wolverines")
for wolverine in wolverines:
#     print(wolverine)
    ab=cv2.imread("C:\\Users\\win_10\\Desktop\\Final Project\\animals\\wolverines\\"+wolverine)
    wolv_array = np.asarray(ab)
    img_from_ar = Image.fromarray(wolv_array, 'RGB')
    resized_image = img_from_ar.resize((50, 50))
    data.append(np.array(resized_image))
    labels.append(3)
    
```

```python
animals=np.array(data)
labels=np.array(labels)
```

```python
np.save("animals",animals)
np.save("labels",labels)
```

```python
s=np.arange(animals.shape[0])
np.random.shuffle(s)
animals=animals[s]
labels=labels[s]
```

```python
num_classes=len(np.unique(labels))
data_length=len(animals)
```

```python
(x_train,x_test)=animals[(int)(0.1*data_length):],animals[:(int)(0.1*data_length)]
x_train = x_train.astype('float32')/255
x_test = x_test.astype('float32')/255
train_length=len(x_train)
test_length=len(x_test)
```

```python
(y_train,y_test)=labels[(int)(0.1*data_length):],labels[:(int)(0.1*data_length)]
```

```python
import keras
from keras.utils import np_utils
#One hot encoding
y_train=keras.utils.to_categorical(y_train,num_classes)
y_test=keras.utils.to_categorical(y_test,num_classes)
```

```python
# import sequential model and all the required layers
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Dense,Flatten,Dropout
#make model
model=Sequential()
model.add(Conv2D(filters=16,kernel_size=2,padding="same",activation="relu",input_shape=(50,50,3)))
model.add(MaxPooling2D(pool_size=2))
model.add(Conv2D(filters=32,kernel_size=2,padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=2))
model.add(Conv2D(filters=64,kernel_size=2,padding="same",activation="relu"))
model.add(MaxPooling2D(pool_size=2))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(500,activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(4,activation="softmax"))
model.summary()
```

```python
# compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', 
                  metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,batch_size=50
          ,epochs=100,verbose=1)
```

```python
score = model.evaluate(x_test, y_test, verbose=1)
print('\n', 'Test accuracy:', score[1])
```

```python
def convert_to_array(img):
    print(img)
    im = cv2.imread(img)
    print(im)
    img = Image.fromarray(im, 'RGB')
    image = img.resize((50, 50))
    return np.array(image)

def get_animal_name(label):
    if label==0:
        return "cat"
    if label==1:
        return "bat"
    if label==2:
        return "spider"
    
def predict_animal(file):
    print("Predicting .................................")
    print(file)
    ar=convert_to_array(file)
    ar=ar/255
    label=1
    a=[]
    a.append(ar)
    a=np.array(a)
    score=model.predict(a,verbose=1)
    print(score)
    label_index=np.argmax(score)
    print(label_index)
    acc=np.max(score)
    animal=get_animal_name(label_index)
    print(animal)
    print("The predicted Animal is a "+animal+" with accuracy =    "+str(acc))
```
